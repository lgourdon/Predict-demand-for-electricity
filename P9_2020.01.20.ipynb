## Diagnostics

### Correlation: Pearson’s Correlation Coefficient
Tests whether two samples have a linear relationship

**Assumptions:**
* Observations in each sample are independent and identically distributed (iid).
* Observations in each sample are normally distributed.
* Observations in each sample have the same variance.


* H0: the two samples are independent.
* H1: there is a dependency between the samples.

source : https://machinelearningmastery.com/statistical-hypothesis-tests-in-python-cheat-sheet/

```python
from scipy.stats import pearsonr
data1 = lr.conso_Mwh
data2 = lr.dju
stat, p = pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

### Normality: Shapiro-Wilk Test
Tests whether a data sample has a Gaussian distribution

**Assumptions:** Observations in each sample are independent and identically distributed (iid)
* H0: the sample has a Gaussian distribution.
* H1: the sample does not have a Gaussian distribution

https://machinelearningmastery.com/statistical-hypothesis-tests-in-python-cheat-shee

```python
# Shapiro-Wilk Normality Test 
from scipy.stats import shapiro
data = lin_reg.resid
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably Gaussian')
else:
    print('Probably not Gaussian')
```

## Normalization of the consumption with the dju dataset

**We have 2 sets of data:**
* A total monthly consumption of electricity over 9 years
* The delta of ºC to reach a constant temperature of 18ºC over 9 years to balance electric consumption due to heating


* Total electric consumption =  heating consumption + other consumption
* Normalized electric consumption 18ºC = Total electric consumption - heating consumption = other consumption


* The linear regression conso vs dju will allow us to estimate the increase of the total consumption for 1 dju

$$ conso = α + 𝛽.dju + ϵ $$
$$ conso.cor = conso - 𝛽.dju = α + ϵ $$

### Calculus

```python
# Correction 
𝛽 = lin_reg.params['dju']
cvl2=pd.DataFrame(lr["conso_Mwh"]- lr['dju']*𝛽)

# Rename
cvl2.rename(columns={0:'conso_cor'},  inplace = True)
cvl2.head(2)
```

```python
# merge
conso=pd.merge(data1, cvl2, left_index=True, right_index=True)
# Rename
conso.rename(columns={'conso_Mwh': 'conso_raw'},  inplace = True)
conso.head(2)
```

# Consumption desaizonalisation – Moving Average (MA)
Considering the curve of the normalized electric consumption, we will chose the **additive model** (constant range around the trend), <br>rather than the multiplicative model (range proportional to the trend)


* **A time serie has 3 componants:** Trend, Seasonality, Residual
* **Aim of desaisonnalization:** Absorb seasonality, conserve the trend and minimize residual variance

## Seasonality - decomposition

```python
# decomposition
conso.index=lr.index.to_timestamp()
s_dec_additive = seasonal_decompose(conso.conso_cor, model = "additive", freq=12)

plt.rcParams["figure.figsize"] = [12,6]
s_dec_additive.plot()
plt.suptitle('Seasonal Decomposition', fontsize=24, color='navy', y=1.03)
plt.xlabel('')
plt.savefig('img/3.1 decomposition.png', dpi=120, bbox_inches='tight')
plt.show()
```

>* **Observed** shows a constant range around the trend => additive model
>* **Trend** is non-linear (flat, ascending or descending)
>* **Seasonal** shows a 12 months periodicity
>* **Residual** seems to be white noise

## Desaisonalization by Moving Average (MA)

```python
# Rolling average
ra=conso.conso_cor.rolling(window=12, center=True).mean()
```

```python
sns.set_style("ticks")
fig, ax = plt.subplots(2, sharex=True, figsize=[28,6])
fig.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)

plt.subplot(2,2,1)
plt.grid(linestyle=':', color='black', linewidth=0.5)
plt.plot(conso.conso_cor, label='Observed', color='royalblue', linewidth=2)
plt.plot(ra, label='MA(12)', color='darkorange', linewidth=2)
plt.title('MA(12) vs Observed', fontsize = 20, color='navy')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(fontsize=13)
plt.ylim(28000,38000)

plt.subplot(2,2,3)
plt.grid(linestyle=':', color='black', linewidth=0.5)
plt.plot(s_dec_additive.trend, label='Trend', color='royalblue', linewidth=2)
plt.plot(ra, label='MA(12)', color='darkorange', linewidth=2)
plt.title('MA(12) vs Trend', fontsize = 20, color='navy')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(fontsize=13)
plt.ylim(31000,33100)

plt.savefig('img/3.2 moving_average.png', dpi=120, bbox_inches='tight')
plt.show()
```

>* The MA(12) desaisonalization is efficient, very close to the Trend obtained during the decomposition of the time-serie
>* The influence of residuals is minimal
>* The model is graphically reliable

## Stationarity

### Augmented Dickey-Fuller Unit Root Test
Tests whether a time series has a unit root, e.g. has a trend or more generally is autoregressive.

**Assumptions :** Observations in are temporally ordered
* H0: a unit root is present (series is non-stationary).
* H1: a unit root is not present (series is stationary).

Source: https://machinelearningmastery.com/statistical-hypothesis-tests-in-python-cheat-sheet/

```python
# Augmented Dickey-Fuller unit root test
data = conso.conso_cor
stat, p, lags, obs, crit, t = adfuller(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably not Stationary')
else:
    print('Probably Stationary')
```

Tests whether a time series is trend stationary or not.

## Autocorrélation

Due to the particular nature of autocorrelation (positive and negative autocorrelation) there is a special test statistic: The Durbin-Watson statistic. <br>Therefore the Durbin-Watson test has a value of the test statistic, but no p-value.

**The Durbin-Watson test statistic works in the following manner:**
* Test statistic value of 0: Perfect positive autocorrelation
* Test statistic value 0f 2: No autocorrelation
* Test statistic value of 4: Perfect negative autocorrelation


If your Durbin-Watson test is perfectly positive autocorrelated you should correct it in order to make the data stationary before doing further analysis.

# Forecast: SARIMA method

Configuring a SARIMA requires selecting hyperparameters for both the trend and seasonal elements of the series.

**Trend Elements**<br>
There are three trend elements that require configuration<br>
They are the same as the ARIMA model<br>
* p: Trend autoregression order (AR)
* d: Trend difference order.
* q: Trend moving average order (MA)

**There are four seasonal elements that are not part of ARIMA that must be configured**<br>
* P: Seasonal autoregressive order.
* D: Seasonal difference order.
* Q: Seasonal moving average order.
* m: The number of time steps for a single seasonal period.


**Together, the notation for an SARIMA model is specified as:**
* SARIMA(p,d,q)(P,D,Q)m

source : https://machinelearningmastery.com/sarima-for-time-series-forecasting-in-python/
